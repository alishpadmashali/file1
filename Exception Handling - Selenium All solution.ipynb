```python
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver=webdriver.Chrome()
driver.get("https://www.amazon.in/")
```

```python
search_product=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search_product.send_keys(input_Value)
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search.click()
```

```python
#lets scrape all the product url 

product_url=[]
start=0
end=3
for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        product_url.append(i.get_attribute('href'))
    next_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    next_button.click()
    time.sleep(3)
```

```python
#Lets scrap the brand name
for url in product_url:
    driver.get(url)
    time.sleep(3)
    try:
        brand=driver.find_element(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table/tbody/tr[1]/td[2]/span')
        Brand.append(brand.text)
    except NoSuchElementException:
        Brand.append('-')
```

```python
#lets Scrape the price

for url in product_url:
    driver.get(url)
    time.sleep(3)
    try:
        price=driver.find_element(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
        
        
        
#Lets scrape the return 

for url in product_url:
    driver.get(url)
    time.sleep(3)
    try:
        returns=driver.find_element(By.XPATH,'//div[@class="a-column a-span12 a-text-center icon-container scrollable-container celwidget"]/span[1]/div[2]/span')
        Return.append(returns.text)
    except NoSuchElementException:
        Return.append('-')
        
        
        
        
#lets check the delivery date


for url in product_url:
    driver.get(url)
    time.sleep(3)
    try:
        delivery=driver.find_element(By.XPATH,'//div[@class="a-spacing-base"]/span[1]/span[1]')
        Expected_delivery.append(delivery.text)
    except NoSuchElementException:
        Expected_delivery.append('-')
```

```python
len(Expected_delivery)
```

```python
len(Return)
```

```python
len(Price)
```

```python
len(Brand)
```

```python
len(product_url)
```

```python
#lets prepare the table using DataFrame
df=pd.DataFrame({'Product URL':product_url,'Brand':Brand,'Price of Product':Price,'Return/Exchange':Return,'Expected Delivery':Expected_delivery})
df
```

```python
csv_data = df.to_csv(index=False)

print(csv_data)
```

```python
search_button=driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div[2]/div/div/div/div[5]/button')
search_button.click()
```

```python
#lets scrap the hostel URL 

HOSTEL_url=[]

url=driver.find_elements(By.XPATH,'//a[@class="property-card-container horizontal"]')
for i in url:
    HOSTEL_url.append(i.get_attribute('href'))
```

```python
len(HOSTEL_url)
```

```python
#lets Scrape the Hostel Name
Hostel_name=[]

for url in HOSTEL_url:
    driver.get(url)
    time.sleep(3)
    try:
        HN=driver.find_element(By.XPATH,'//div[@class="title-2"]')
        Hostel_name.append(HN.text)
    except NoSuchElementException:
        Hostel_name.append('-')
```

```python
len(Hostel_name)
```

```python
print(Hostel_name)
```

```python
#lets Scrape the Location/Address of hostel
Hostel_address=[]

for url in HOSTEL_url:
    driver.get(url)
    time.sleep(3)
    try:
        HA=driver.find_element(By.XPATH,'//button[@class="property-address"]')
        Hostel_address.append(HA.text)
    except NoSuchElementException:
        Hostel_address.append('-')
        
        
        
        
#lets Scrape the description
Description=[]

for url in HOSTEL_url:
    driver.get(url)
    time.sleep(3)
    try:
        Desc=driver.find_element(By.XPATH,'//div[@class="description-container"]/div[1]/div[2]')
        Description.append(Desc.text)
    except NoSuchElementException:
        Description.append('-')       
        

        
#lets Scrape the Rating
Rating=[]

for url in HOSTEL_url:
    driver.get(url)
    time.sleep(3)
    try:
        rate=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[1]')
        Rating.append(rate.text)
    except NoSuchElementException:
        Rating.append('-')
        
        
        
#Lets Scrape the overall Rating       
Overall_rating=[]

for url in HOSTEL_url:
    driver.get(url)
    time.sleep(3)
    try:
        overall=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div[1]/span[1]')
        Overall_rating.append(overall.text)
    except NoSuchElementException:
        Overall_rating.append('-')
        
        
        
#Lets scrape the Total review

Total_review=[]

for url in HOSTEL_url:
    driver.get(url)
    time.sleep(3)
    try:
        total=driver.find_element(By.XPATH,'//div[@class="flex-20"]/div[1]/div[2]/div[2]')
        Total_review.append(total.text.split(' ')[0])
    except NoSuchElementException:
        Total_review.append('-')
        
        
        
        
        
#lets scrape the private room price

Private=[]


for url in HOSTEL_url:
    driver.get(url)
    time.sleep(3)
    try:
        private=driver.find_element(By.XPATH,'//div[@class="rate-container bordered"]/div[1]/div[2]/div[1]')
        Private.append(private.text)
    except NoSuchElementException:
        Private.append('-')
```
